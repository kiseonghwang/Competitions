```python
import pandas as pd

target = pd.read_csv("./kaist_target.csv")
```

```python
import numpy as np

sample_count = 277
time_interval = 0.04  # 40ms   

time = np.array([i * time_interval for i in range(sample_count)])

print(time)
X_time = time
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, LSTM, concatenate, Dense, Dropout
from tensorflow.keras.models import Model

def create_model(image_shape = [240, 360, 3], time_steps = 10, num_features = 16):
    image_input = Input(shape=[240, 360, 3])  #   
    conv1 = Conv2D(32, (3, 3), activation='selu')(image_input)
    pool1 = MaxPooling2D((2, 2))(conv1)
    conv2 = Conv2D(64, (3, 3), activation='selu')(pool1)
    pool2 = MaxPooling2D((2, 2))(conv2)
    flat = Flatten()(pool2)

    time_input = Input(shape=(time_steps, num_features))  #   
    lstm1 = LSTM(32, return_sequences=True)(time_input)
    lstm2 = LSTM(16)(lstm1)

    merged = concatenate([flat, lstm2])  # CNN RNN 
    dense1 = Dense(128, activation='selu')(merged)
    dropout1 = Dropout(0.5)(dense1)
    dense2 = Dense(64, activation='selu')(dropout1)
    dropout2 = Dropout(0.5)(dense2)
    output = Dense(16, activation='linear')(dropout2)  #   linear  

    model = Model(inputs=[image_input, time_input], outputs=output)  #  
    model.compile(optimizer='nadam', loss='mse')  #     loss    

    return model
#  
model_instance = create_model()
```

```python
full_X_train, X_test, full_y_train, y_test = train_test_split(resized_images_data, target, test_size=0.1, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(full_X_train, full_y_train, test_size=0.1, random_state=42)
full_X_train_time, X_test_time, full_X_train_time, y_test_time = train_test_split(X_time, time, test_size=0.1, random_state=42)
X_train_time, y_train_time, X_valid_time,y_valid_time = train_test_split(full_X_train_time, full_X_train_time, test_size=0.1, random_state=42)
```

```python
from tensorflow.keras.callbacks import TensorBoard

import os
root_logdir = os.path.join(os.curdir, "_mylogs7")
def get_run_logdir():
    import time
    run_id = time.strftime("run_%Y_%m_%d-H_%M_%S")
    return os.path.join(root_logdir, run_id)

run_logdir = get_run_logdir()
tensorboard_cb = keras.callbacks.TensorBoard(run_logdir)
```

```python
early_stooping_cb = keras.callbacks.EarlyStopping(patience=20,
                                                  restore_best_weights=True)

checkpoint_cb = keras.callbacks.ModelCheckpoint("save_my_keras_model.h5",
                                                save_best_only=True)
```

```python
num_samples = 224
time_steps = 10
num_features = 16
seed_value = 42
np.random.seed(seed_value)
#    
import numpy as np
X_train_time_example = np.random.rand(num_samples * time_steps * num_features)

#  
X_train_time_reshaped = X_train_time_example.reshape(num_samples, time_steps, num_features)

num_valid_samples = X_valid.shape[0]

X_valid_time_example = np.random.rand(num_valid_samples * time_steps * num_features)
X_valid_time_reshaped = X_valid_time_example.reshape(num_valid_samples, time_steps, num_features)

num_test_samples = X_test.shape[0]

X_test_time_example = np.random.rand(num_test_samples * time_steps * num_features)
X_test_time_reshaped = X_test_time_example.reshape(num_test_samples, time_steps, num_features)


#        .
history = model_instance.fit([X_train, X_train_time_reshaped], y_train, batch_size=32, epochs=1000000000, validation_data=([X_valid, X_valid_time_reshaped], y_valid),
                            callbacks=[tensorboard_cb, early_stooping_cb])
```
