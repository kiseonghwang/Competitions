```python
# 3. 소스 링크 수집 URL

urls = [
    "https://kr.hotels.com/go/south-korea/kr-best-things-to-do-jeju",
    "https://kr.hotels.com/go/south-korea/kr-best-things-to-do-this-winter-on-jeju-island",
    "https://kr.hotels.com/go/south-korea/kr-best-jeju-nightlife",
    "https://kr.hotels.com/go/south-korea/best-jeju-shopping",
    "https://kr.hotels.com/go/south-korea/kr-best-jeju-island-dining-area"
]
```

```python
# 4. URL loader 생성-데이터 읽어오기
from langchain.document_loaders import UnstructuredURLLoader

loader = UnstructuredURLLoader(urls=urls)
data = loader.load()
print(data)
```

```python
# 5. Text Split(GPT API Token Limit)

from langchain.text_splitter import TokenTextSplitter

text_splitter = TokenTextSplitter(chunk_size=1000, chunk_overlap=50)
doc = text_splitter.split_documents(data)
print(doc)
```

```python
# 6. Vetor Store에 텍스트 벡터값 저장(검색가능함)

from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma

embeddings = OpenAIEmbeddings(
    model='text-embedding-ada-002') # 사람이 인식하는 문자 -> 컴튜터 인식 벡터 모델
vectorstore = Chroma.from_documents(doc, embeddings) # 문자 -> 벡터 변환해서 벡터 스토리지에 저
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k":5})
# 리트리버 벡터 문서 -> LLM 모델 인식할 수 있는 텍스트 문서로 반환/조회
```

```python
# 7. LLM 연결

from langchain.chat_models import ChatOpenAI

chat_llm = ChatOpenAI(
    model_name="gpt-3.5-turbo",
    temperature=0
)
```

```python
# 8. Chain 연결

from langchain.chains import RetrievalQA

qa = RetrievalQA.from_chain_type(
    llm=chat_llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True)
```

```python
# 9. 질의
query = "제주도 여행에서 꼭 들려야 하는 곳을 5곳 추천해줘"
result = qa({"query": query})

print(result["result"], result["source_documents"][0].metadata)
```
