```python
# 2. 데이터 업로드
# Finetuning_chat_data.jsonl
# Finetuning_chat.jsonl
training_file_name = "/content/electronics_chat_data.jsonl"

training_response = openai.File.create(
    file=open(training_file_name, "rb"), purpose="fine-tune"
)

training_file_id = training_response["id"]
print(training_file_id)
```

```python
# 3. Finetune 작업 생성(약 10분)

response = openai.FineTuningJob.create(
    training_file=training_file_id,
    model="gpt-3.5-turbo",
    suffix="chat_giseong07"
)

job_id = response["id"]

print(job_id)
```

```python
# 모델 학습 진행상황 체크

model_response = openai.FineTuningJob.retrieve(job_id)
fine_tuned_model_status = model_response["status"]
print(fine_tuned_model_status)
```

```python
# 모델 id 조회

model_response = openai.FineTuningJob.retrieve(job_id)
fine_tuned_model_id = model_response["id"]
print(fine_tuned_model_id)
print(model_response["fine_tuned_model"])
```
