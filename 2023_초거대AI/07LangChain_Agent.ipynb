```python
# 2. 패키지 설정

import os
import openai
from langchain.agents import initialize_agent, load_tools
from langchain.agents import AgentType
from langchain.chat_models import ChatOpenAI
```

```python
# 3. OpenAI 환경설정

import os

# fine-tuning, lagchain 사용시 반드시 환경변수 값 사용!!
os.environ["OPENAI_API_KEY"] = "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"
os.environ["SERPAPI_API_KEY"] = "4d5d090f2d1dcf01e8408e2cc2fd48e6f8770505bd6935c817b0d9539fa734a0" # 구글 검색을 할 수 있게 해주는 키 값

openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
# 4. Chat 모델 생성

chat_llm = ChatOpenAI(
    model_name="gpt-3.5-turbo",
    temperature=0.5
)
```

```python
# 5. Tool 생성 및 Agent 연결

tools_name = ["serpapi"] # tool은 여러개를 쓸 수 있지만 필요한 것만 사용하는 것이 좋다.
tools = load_tools(tools_name)

agent = initialize_agent(tools,
                         chat_llm,
                         agent=AgentType.CHAT_ZERO_SHOT_REACT_DESCRIPTION,
                         verbose=True) # agent가 액션-관찰-생각하는 과정을 보여달라
```

```python
# 6. Agent 실행

agent.run("2023년 아이폰 최신 모델을 한글로 알려줘")
```
