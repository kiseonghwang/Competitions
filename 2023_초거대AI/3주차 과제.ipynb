# 초거대 AI Finetuning 실습

### 경북A1/황기성/러닝머신

* OpenAI 패키지 설치
* OpenAI 환경 설정
* GPT chat 모델 Finetuning

```python
training_file_name = "/content/Finetuning_chat.jsonl"

training_response = openai.File.create(
    file=open(training_file_name, "rb"), purpose="fine-tune"
)

training_file_id = training_response["id"]
print(training_file_id)
```

```python
# 3. Finetune 작업 생성(약 10분)

response = openai.FineTuningJob.create(
    training_file=training_file_id,
    model="gpt-3.5-turbo",
    suffix="chat_giseong07"
)

job_id = response["id"]

print(job_id)
```

```python
job_id = response["id"]

print(job_id)
```

```python
# 모델 학습 진행상황 체크

model_response = openai.FineTuningJob.retrieve(job_id)
fine_tuned_model_status = model_response["status"]
print(fine_tuned_model_status)
```

```python
# 모델 id 조회

model_response = openai.FineTuningJob.retrieve(job_id)
fine_tuned_model_id = model_response["id"]
print(fine_tuned_model_id)
print(model_response["fine_tuned_model"])
```

```python
# 파인튜닝 전 일반 gpt 3.5 모델 (틀린 정보 혹은 없는 챔피언이라고 나옴)

response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "롤 캐릭터 이름을 알려주면 포지션, 스토리를 알려주는 친절한 도우미"},
        {"role":"user", "content":"제리"}
    ]
)

print(response.choices[0].message.content)
```

```python
# 4. 리그오브 레전드 게임 캐릭터들과 아이템의 정보를 담아 파인튜닝 한 후 모델 (정확하지 않지만 어느정도 맞는 정보 제공 약 200개)

response = openai.ChatCompletion.create(
    model="ft:gpt-3.5-turbo-0613:swhub-ai:chat-giseong07:7yZbJhQG",
    messages=[
        {"role": "system", "content": "롤 캐릭터 이름을 알려주면 포지션, 스토리를 알려주는 친절한 도우미"},
        {"role":"user", "content":"제리"}
    ]
)

print(response.choices[0].message.content)
```
