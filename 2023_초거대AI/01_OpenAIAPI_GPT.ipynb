```python
# 3. Completion API 호출

response = openai.Completion.create( # 뒷 문장 자동 완성
    model="text-davinci-003", # 사용할 모델의 이름
    prompt="한글 영화 스토리 작성해줘: 그녀는 긴 머리카락을 바람에 휘날리며",
    temperature=1, # 결과물의 창의성 조절
    max_tokens=300,
    stop=['.'],
    n=30, # 한번에 4개의 결과물을 만들어라
    stream=True # 기본값은 Flase이고 True로 지정하면 결과 문장의 생성된 토큰을 실시간으로 받아온다.
)
for chunk in response:
  print(chunk.choices[0].text)
# response["choices"][0]["text"]
```

```python
for choice in response["choices"]:
  print(choice['text'])
```

```python
# 4. Chat Completion

response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    max_tokens=300,
    messages=[
        {"role": "system", "content": "항상 문장 끝에 이모지를 붙이는 친절한 수학 선생님"},
        {"role": "user", "content": "선생님 안녕하세요?"},
        {"role": "assistant", "content": "안녕하세요, 학생 여러분. 반갑습니다.🤗"},
        {"role": "assistant", "content": "적분이란 미분의 반대되는 개념으로, 다양한 분야에서 활용됩니다.🤗"},
        {"role": "user", "content": "적분은 어디에 활용되나요??"}
    ],
    stream=True
)
# print(response.choices[0].message.content)
for chunk in response:
  print(chunk.choices[0].delta.content)
```
